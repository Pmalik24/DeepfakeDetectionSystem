```python
import re
import matplotlib.pyplot as plt

def extract_loss_accuracy_from_logs(file_path):
    """
    Extracts training and validation loss and accuracy from log files.

    Args:
        file_path (str): Path to the log file.

    Returns:
        tuple: Lists of training loss, validation loss, training accuracy, validation accuracy.
    """
    train_loss = []
    val_loss = []
    train_accuracy = []
    val_accuracy = []
    with open(file_path, 'r') as file:
        for line in file:
            # Extract training and validation loss
            train_loss_match = re.search(r"Train Loss: ([0-9.]+)", line)
            val_loss_match = re.search(r"Val Loss: ([0-9.]+)", line)
            train_acc_match = re.search(r"Train Accuracy: ([0-9.]+)", line)
            val_acc_match = re.search(r"Val Accuracy: ([0-9.]+)", line)
            
            if train_loss_match:
                train_loss.append(float(train_loss_match.group(1)))
            if val_loss_match:
                val_loss.append(float(val_loss_match.group(1)))
            if train_acc_match:
                train_accuracy.append(float(train_acc_match.group(1)))
            if val_acc_match:
                val_accuracy.append(float(val_acc_match.group(1)))
    
    return train_loss, val_loss, train_accuracy, val_accuracy

```

```python
# File paths for the log files
swin_log_path = 'swin_baby_log_2.txt'
vit_log_path = 'best_model_training_log.txt'
cnn_log_path = 'cnn_training_log.txt'

# Extract data for SWIN, ViT & CNN
swin_train_loss, swin_val_loss, swin_train_accuracy, swin_val_accuracy = extract_loss_accuracy_from_logs(swin_log_path)
vit_train_loss, vit_val_loss, vit_train_accuracy, vit_val_accuracy = extract_loss_accuracy_from_logs(vit_log_path)
cnn_train_loss, cnn_val_loss, cnn_train_accuracy, cnn_val_accuracy = extract_loss_accuracy_from_logs(cnn_log_path)

# Generate epochs for plotting
swin_epochs = range(1, len(swin_train_loss) + 1)
vit_epochs = range(1, len(vit_train_loss) + 1)
cnn_epochs = range(1, len(cnn_train_loss) + 1)
```

```python

# Plot SWIN: Training vs Validation Loss
plt.figure(figsize=(10, 6))
plt.plot(swin_epochs, swin_train_loss, label="Training Loss", linestyle="--", marker="o", color="blue")
plt.plot(swin_epochs, swin_val_loss, label="Validation Loss", linestyle="-", marker="o", color="orange")
plt.title("SWIN: Training vs Validation Loss", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot SWIN: Training vs Validation Accuracy
plt.figure(figsize=(10, 6))
plt.plot(swin_epochs, swin_train_accuracy, label="Training Accuracy", linestyle="--", marker="o", color="blue")
plt.plot(swin_epochs, swin_val_accuracy, label="Validation Accuracy", linestyle="-", marker="o", color="orange")
plt.title("SWIN: Training vs Validation Accuracy", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ViT: Training vs Validation Loss
plt.figure(figsize=(10, 6))
plt.plot(vit_epochs, vit_train_loss, label="Training Loss", linestyle="--", marker="x", color="green")
plt.plot(vit_epochs, vit_val_loss, label="Validation Loss", linestyle="-", marker="x", color="red")
plt.title("ViT: Training vs Validation Loss", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot ViT: Training vs Validation Accuracy
plt.figure(figsize=(10, 6))
plt.plot(vit_epochs, vit_train_accuracy, label="Training Accuracy", linestyle="--", marker="x", color="green")
plt.plot(vit_epochs, vit_val_accuracy, label="Validation Accuracy", linestyle="-", marker="x", color="red")
plt.title("ViT: Training vs Validation Accuracy", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot CNN: Training vs Validation Loss
plt.figure(figsize=(10, 6))
plt.plot(cnn_epochs, cnn_train_loss, label="Training Loss", linestyle="--", marker="o", color="purple")
plt.plot(cnn_epochs, cnn_val_loss, label="Validation Loss", linestyle="-", marker="o", color="cyan")
plt.title("CNN: Training vs Validation Loss", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Loss", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot CNN: Training vs Validation Accuracy
plt.figure(figsize=(10, 6))
plt.plot(cnn_epochs, cnn_train_accuracy, label="Training Accuracy", linestyle="--", marker="o", color="purple")
plt.plot(cnn_epochs, cnn_val_accuracy, label="Validation Accuracy", linestyle="-", marker="o", color="cyan")
plt.title("CNN: Training vs Validation Accuracy", fontsize=16)
plt.xlabel("Epochs", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
